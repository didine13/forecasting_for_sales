```python
def unit_sales(x):
    tmp = 0
    if x[2] == 0 :
        tmp =  x+ x*0.3
    else :
        tmp = x
    #df['stock'] = tmp
    return tmp

stock = pd.DataFrame(sales_items[['store_nbr', 'family', 'unit_sales', 'perishable', 'item_nbr']].groupby(['store_nbr','family', 'perishable', 'item_nbr']).count()['unit_sales']).reset_index()
stock
```

```python
def apply_restock_days(df):
    dict_map = {'BREAD/BAKERY': 5,
               'DAIRY': 7,
               'DELI': 3,
               'EGGS': 25,
               'MEATS': 3,
               'POULTRY': 3,
               'PREPARED FOODS': 3,
               'PRODUCE': 5,
               'SEAFOOD': 2,
               'MAGAZINES': 10,
               }
    df['replenishment_period'] = df['family'].map(dict_map).fillna(30)
    return df
```

```python
items = apply_restock_days(items)
```
